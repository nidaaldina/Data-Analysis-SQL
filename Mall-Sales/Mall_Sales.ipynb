```python
import pandas as pd
```

```python
# membaca file sales data

sales = pd.read_csv('sales_data.csv')
sales.head(15)
```

```python
# mengecek informasi data sales

sales.info()
```

```python
print('Jumlah data duplikat:', sales.duplicated().sum())
```

```python
# mengecek jumlah nilai yang unique pada kolom invoice_no dan customer_id
invoice_no_unique = sales['invoice_no'].nunique()
print('Jumlah nilai unique pada kolom invoice_no:', invoice_no_unique)

customer_id_unique = sales['customer_id'].nunique()
print('Jumlah nilai unique pada kolom customer_id:', customer_id_unique)
```

```python
sales.describe()
```

```python
# Ubah tipe data menjadi datetime
sales['invoice_date'] = pd.to_datetime(sales['invoice_date'], format='%d-%m-%Y')

print(sales.dtypes)
```

```python
sales.head()
```

```python
# Memecah kolom "Date" menjadi "Year" dan "Month"
sales['year'] = pd.to_datetime(sales['invoice_date']).dt.year
sales['month'] = pd.to_datetime(sales['invoice_date']).dt.month
```

```python
# Mengubah angka bulan ke nama bulan singkat
month_rename = {
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr',
    5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug',
    9: 'Sep', 10: 'Okt', 11: 'Nov', 12: 'Des'
}

sales['month_name'] = sales['month'].map(month_rename)

sales.head()
```

```python
sales.to_csv('salesdata.csv')
```

```python
# membaca data customer

customer = pd.read_csv('customer_data.csv')
customer.head(15)
```

```python
# mengecek informasi customer
customer.info()
```

```python
# memastikan apakah ada data duplikasi
print('Jumlah data duplikasi: ', customer.duplicated().sum())
```

```python
# mengecek nilai unique pada customer_id
customer_id_unique = customer['customer_id'].nunique()
print('Jumlah data unique pada customer_id:', customer_id_unique)
```

```python
customer.describe()
```

```python
# mengecek jumlah missing value pada kolom age
print(customer['age'].isnull().sum())
```

```python
# menghapus data yang memiliki missing value
customer.dropna(inplace=True)
```

```python
customer.isnull().sum()
```

```python
# Ubah tipe data kolom 'age' menjadi integer
customer['age'] = customer['age'].astype(int)

# Periksa tipe data setelah perubahan
print(customer.dtypes)
```

```python
customer.head()
```

```python
customer.info()
```

```python
customer.to_csv('customerdata.csv')
```
